```python
import fitz  # PyMuPDF
import re

# Open PDF
doc = fitz.open("D:\\C\\downloads1\\combined act.pdf")

# List of words to highlight (Yellow) with exact word boundaries
search_words = [
    r"\bRUPEES\b", r"\bRS\b", r"\bMORE\b", r"\bLESS\b", r"\bAMOUNT\b", r"\bPAID\b", r"\bAPPLICABLE\b",
    r"\bRATE OF\b", r"\bRATE\b", r"\bQUALIFIED\b", r"\bQUALIFY\b", r"\bQUALIFICATION\b",
    r"\bDAY\b", r"\bWEEK\b", r"\bMONTH\b", r"\bYEAR\b", r"\bDURATION\b", r"\bALLOWANCE\b", r"\bMEDIATE\b",
    r"\bFORM\b", r"\bFORM NO\b", r"\bSHARE\b", r"\bCONTRIBUTION\b", r"\bAGE\b", r"\bMINIMUM\b", r"\bMAXIMUM\b",
    r"\bNIGHT\b", r"\bOBLIGATION\b", r"\bPERIOD\b", r"%", r"\bPERCENTAGE\b", r"\bPER CENT\b", r"\bSANITION\b",
    r"\bCOMPETENT\b", r"\bNOTICE\b", r"\bDEFAULT\b", r"\bINCENTIVE\b", r"\bCOMMITTEE\b", r"\bUNDER\b", r"\bSCHEDULED\b",
    r"\bSCHEDULE\b", r"\bSUB-CLAUSE\b", r"\bSUB CLAUSE\b", r"\bPROHIBIT\b", r"\bFULL\b", r"\bSALARY\b", r"\bNOT INCLUDE\b", r"\bNOT COMPLETE\b",
    r"\bHUNDRED\b", r"\bTHOUSAND\b", r"\bLAKH\b", r"\bPOWER\b", r"\bEVERY\b", r"\bGOVERN\b", r"\bCOMPULSORY\b", r"\bCOMPULSION\b",
    r"\bAPPOINTMENT\b", r"\bHOUR\b", r"\bPUBLIC\b", r"\bCOUNCIL\b", r"\bOMIT\b", r"\bOMITTED\b", r"\bREPEAL\b", r"\bATTAIN\b",
    r"\bTERMINATE\b", r"\bMUST\b",r"\bFULLY\b"
]

# Regex for numbers & currency (only Rs / ₹ / rupees forms, not part of other words)
number_pattern = re.compile(
    r"\b(?:Rs\.?|₹)\s?\d{1,3}(?:,\d{3})*(?:\.\d+)?\b"       # Rs.5000 / ₹200
    r"|\b\d+(?:\.\d+)?\s?(?:rupees?|%|percent|per cent|year|month|day|hour|hours)\b",  
    re.IGNORECASE
)

for page in doc:
    text = page.get_text("text")

    # --- Highlight fixed words (Yellow) ---
    for pattern in search_words:
        for match in re.finditer(pattern, text, flags=re.IGNORECASE):
            found_text = match.group()
            instances = page.search_for(found_text, flags=1)  # case-insensitive
            for inst in instances:
                highlight = page.add_highlight_annot(inst)
                highlight.set_colors(stroke=(1, 1, 0))  # Yellow
                highlight.update()

    # --- Highlight numbers & currency (Light Green) ---
    for match in number_pattern.finditer(text):
        found_text = match.group()
        number_instances = page.search_for(found_text, flags=1)
        for inst in number_instances:
            highlight = page.add_highlight_annot(inst)
            highlight.set_colors(stroke=(0.6, 1, 0.6))  # Light Green
            highlight.update()

# Save final PDF
doc.save("D:\\C\\downloads1\\highlighted2.pdf")
```
